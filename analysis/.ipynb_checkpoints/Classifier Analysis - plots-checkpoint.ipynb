```r
options(repr.plot.width=4, repr.plot.height=4)
library(ggplot2)

df = read.csv("unit_impact.csv")
df$model <- as.factor(df$model)

ggplot(data=df, aes(x=ccg, y=lm, color=model)) + geom_point(size=0.3) + 
theme_minimal() + 
theme(axis.text.x=element_text(size=12), axis.text.y=element_text(size=12),
          axis.title.y=element_text(size=12),
          strip.text=element_text(size=12, angle=0), legend.text=element_text(size=8), 
          panel.grid.minor=element_blank(),
          strip.text.y=element_text(angle=0), legend.position="None") +
labs(x="Unit EV on Supertagging Task", y="Unit EV on Word Prediction Task")

ggsave("plots/aug_impact_scatter.png", width=4, height=4)
```
