# Franck 2002

```r
franck_df <- read.csv('../evalsets/Franck2002/items.csv.reformat')

franck_df <- tidyr::separate(franck_df, condition, c("subj", "inter", "local"), sep=1:3)
franck_df$sample <- 100 - franck_df$sample
franck_df$max <- 100 - franck_df$max
franck_df

franck_lstm_df = subset(franck_df, substr(franck_df$model, 0,4) == "lstm")

franck_rnng_df = subset(franck_df, substr(franck_df$model, 0,4) == "rnng")
```

```r
plt <- ggplot(franck_lstm_df, aes(x=inter, fill=local, y=max)) +
geom_bar(stat="summary", position="dodge", fun.y = "mean") + 
stat_summary(fun.data = mean_se, geom = "errorbar", position="dodge") +
labs(title="LSTM MaxProb results for Franck et al. (2002)", x="Intermediate ('high') attractor #", 
     fill="Local ('low') attractor #", y="% attraction errors")  +
facet_grid(~ subj) +
coord_cartesian(ylim=c(0,50))

plt

ggsave("Franck_lstm_max.png")

plt <- ggplot(franck_rnng_df, aes(x=inter, fill=local, y=max)) +
geom_bar(stat="summary", position="dodge", fun.y = "mean") + 
stat_summary(fun.data = mean_se, geom = "errorbar", position="dodge") +
labs(title="RNNG MaxProb results for Franck et al. (2002)", x="Intermediate ('high') attractor #", 
     fill="Local ('low') attractor #", y="% attraction errors") +
facet_grid(~ subj) +
coord_cartesian(ylim=c(0,50))

plt

ggsave("Franck_rnng_max.png")
```

```r
plt <- ggplot(franck_lstm_df, aes(x=inter, fill=local, y=sample)) + 
geom_bar(stat="summary", position="dodge", fun.y = "mean") + 
stat_summary(fun.data = mean_se, geom = "errorbar", position="dodge") +
labs(title="LSTM Sampling results for Franck et al. (2002)", x="Intermediate ('high') attractor #", 
     fill="Local ('low') attractor #", y="% attraction errors") +
facet_grid( ~subj, labeller = label_both) +
coord_cartesian(ylim=c(0,60))

plt

ggsave("Franck_lstm.png")
```

```r
ggplot(franck_rnng_df, aes(x=inter, fill=local, y=sample)) + 
geom_bar(stat="summary", position="dodge", fun.y = "mean") + 
labs(title="RNNG Sampling results for Franck et al. (2002)", x="Intermediate ('high') attractor #", 
     fill="Local ('low') attractor #", y="% Attraction Errors") +
stat_summary(fun.data = mean_se, geom = "errorbar", position="dodge") +
facet_grid( ~subj, labeller = label_both) +
coord_cartesian(ylim=c(0,60))


ggsave("Franck_rnng.png")
```

```r
ggplot(franck_rnng_df, aes(x=inter, fill=inter, y=sample)) + 
geom_bar(stat="summary", position="dodge", fun.y = "mean") + 
stat_summary(fun.data = mean_se, geom = "errorbar", position="dodge") +
facet_grid( ~model, labeller = label_both) +
labs(x="Intermediate Attractor #", y="% Attraction Errors")+
theme(legend.position = "none") +
coord_cartesian(ylim=c(0,50))

ggsave("Franck_rnng_bymodel_inter.png")

ggplot(franck_rnng_df, aes(x=local, fill=local, y=sample)) + 
geom_bar(stat="summary", position="dodge", fun.y = "mean") + 
stat_summary(fun.data = mean_se, geom = "errorbar", position="dodge") +
facet_grid( ~model, labeller = label_both) +
labs(x="Local Attractor #", y="% Attraction Errors") +
theme(legend.position = "none") +
coord_cartesian(ylim=c(0,50))

ggsave("Franck_rnng_bymodel_local.png")
```

```r
franck_lm <- lmer(sample ~ inter * local + (inter * local | item) + (inter * local | model), data=franck_lstm_df)

lstm_eff <- summary(franck_lm)$coeff[c("interS", "localS"), 1]
lstm_se <- summary(franck_lm)$coeff[c("interS", "localS"), 2]

lstm_eff
```

```r
franck_lm <- lmer(sample ~ inter * local + (inter * local | item) + (inter * local | model), data=franck_rnng_df)

rnng_eff <- summary(franck_lm)$coeff[c("interS", "localS"), 1]
rnng_se <- summary(franck_lm)$coeff[c("interS", "localS"), 2]

rnng_eff

eff = c(lstm_eff, rnng_eff)
eff

se = c(lstm_se, rnng_se)
```

```r
eff_df <- data.frame(model=c("LSTM", "LSTM", "RNNG", "RNNG"), attr=c("intermediate", "local", "intermediate", "local"), eff=eff, se=se)

eff_df

ggplot(eff_df, aes(y=-eff, x = attr, fill=attr)) + 
facet_grid( ~ model) +
geom_bar(stat="identity", position="dodge") +
geom_errorbar(aes(ymax=-eff+se, ymin=-eff-se), width=0.5) +
theme(legend.position = "none") +
coord_cartesian(ylim=c(-3, 10)) +
labs(y=expression(beta), x="Attractor Position") 

ggsave("Franck_effectsize.png")
```

# Bock 1992

```r
b92_df <- read.csv("../evalsets/Bock1992/items.csv.reformat")
b92_df <- tidyr::separate(b92_df, condition, c("nums","clause"), sep="_")
b92_df <- tidyr::separate(b92_df, nums, c("subj", "attr"), sep=1:2)

b92_df$type <- substr(b92_df$model, 0, 4)
b92_df$match <- b92_df$subj == b92_df$attr

b92_df$sample <- 100 - b92_df$sample

b92_df$max <- 100 - b92_df$max
b92_df
```

```r
ggplot(b92_df, aes(x=clause, fill=match, y=sample)) + 
geom_bar(stat="summary", position="dodge", fun.y = "mean") + 
stat_summary(fun.data = mean_se, geom = "errorbar", position="dodge") +
facet_grid( ~type, labeller = label_both)
```

```r
b92_lm <- lmer(data=subset(b92_df, type=="lstm"), sample ~ match * clause + (match * clause || item) + (match * clause || model))
summary(b92_lm)

b92_lm <- lmer(data=subset(b92_df, type=="rnng"), sample ~ match * clause + (match * clause || item) + (match * clause || model))
summary(b92_lm)
```

```r
b92_lm <- lmer(data=subset(b92_df, type=="rnng" & clause=="rc"), sample ~ match + (1 | item) + (1 | model))
rnng_rc_eff <- -summary(b92_lm)$coeff[2,1]
rnng_rc_se <- summary(b92_lm)$coeff[2,2]

b92_lm <- lmer(data=subset(b92_df, type=="lstm" & clause=="rc"), sample ~ match + (1 | item) + (1 | model))
lstm_rc_eff <- -summary(b92_lm)$coeff[2,1]
lstm_rc_se <- summary(b92_lm)$coeff[2,2]

b92_lm <- lmer(data=subset(b92_df, type=="rnng" & clause=="prop"), sample ~ match + (1 |item) + (1 | model))
rnng_pp_eff <- -summary(b92_lm)$coeff[2,1]
rnng_pp_se <- summary(b92_lm)$coeff[2,2]

b92_lm <- lmer(data=subset(b92_df, type=="lstm" & clause=="prop"), sample ~ match + (1 | item) + (1 | model))
lstm_pp_eff <- -summary(b92_lm)$coeff[2,1]
lstm_pp_se <- summary(b92_lm)$coeff[2,2]

b92_effs <- data.frame(type=c("rnng", "lstm", "rnng", "lstm"), clause=c("rc", "rc", "pp", "pp"), 
                       eff=c(rnng_rc_eff, lstm_rc_eff, rnng_pp_eff, lstm_pp_eff),
                       se=c(rnng_rc_se, lstm_rc_se, rnng_pp_se, lstm_pp_se))
b92_effs
```

```r
ggplot(b92_effs, aes(y=eff, x = clause, fill=clause)) + 
facet_grid( ~ type) +
geom_bar(stat="identity", position="dodge") +
geom_errorbar(aes(ymax=se+eff, ymin=-se+eff), width=0.5) +
theme(legend.position = "none") +
labs(x = "Clause Type", y = expression(paste("Size of Attraction Effect (", beta, ")")))

expression(beta)
b92_effs$eff
```
